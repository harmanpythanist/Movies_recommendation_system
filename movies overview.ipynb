# Creating Function to reccoment movies based on upper code

```python
def recomendation(title, simi, indices):
    title = title.lower()
    try:
        
        idx = indices[title]
        simi_score = list(enumerate(simi[idx]))
        sim_score = sorted(simi_score, key=lambda x:x[1], reverse=True)
    
        scores_for_top = sim_score[1:11]
    #print('scores for top: ')
    #print(scores_for_top)
        movies = [i[0] for i in scores_for_top]
    #print('movies: ')
    #print(movies)
        print('Based on the movie ' + title + ', you can watch: ')
        print('')
        for i in movies:
            print(file2['title'][i])
            
    except:
        print('no movi found')
    #print('I think you should try: ')
    #return file2['title'].iloc[movies]
```
